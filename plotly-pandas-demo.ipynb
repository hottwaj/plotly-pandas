```python
from plotly.offline import init_notebook_mode, iplot
init_notebook_mode(connected=False)
```

```python
from IPython.display import display
import plotly_pandas as ply_pd
```

```python
data = [ 
    {'x': [1999, 2000, 2001, 2002], 'y': [10, 15, 13, 17], 'type': 'scatter'}, 
    {'x': [1999, 2000, 2001, 2002], 'y': [16, 5, 11, 9], 'type': 'scatter'}
]
 
layout = { 
    'title': 'Sales Growth', 
    'xaxis': {'title': 'Year', 'showgrid': False, 'zeroline': False}, 
    'yaxis': {'title': 'Percent', 'showline': False},
    'width': 600,
    'height': 400
}
```

```python
iplot({'data': data, 'layout': layout})
```

```python
p = ply_pd.PlotlyChartBundle(data_layout = {'data': data, 'layout': layout}, 
                             config = ply_pd.default_config,
                             width = 600, 
                             height = 400)
p
```

```python
import pandas
import datetime
df = pandas.DataFrame({'x': [4,6,3,7,9,5,3]}, index = [datetime.date(2017, 1, i+1) for i in range(7)])
#df.index = pandas.to_datetime(df.index)
ply_pd.chart(df)
```

```python
ply_pd.charts_table([
    ply_pd.chart(df),
    ply_pd.chart(df * 2)
], cols = 2)
```

```python
ply_pd.charts_table([
    df,
    df.style.format("{:.2f}"),  #n.b. style requires jinja2 package to be installed
    ply_pd.chart(df)
], cols = 3)
```
